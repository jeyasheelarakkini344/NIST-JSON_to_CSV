```python
import pandas as pd
original_dataset=pd.read_json('https://raw.githubusercontent.com/usnistgov/SelfishMiningSim/master/all_results.json')
original_dataset.head(5)
```

```python
df=original_dataset
diffparam=df['difficulty_parameters'].apply(pd.Series)
list_df=[]
for i in range(len(df)):
    df1=df['results'].iloc[i]
    data=pd.DataFrame(df1)
    data.insert(0,'daa',df['daa'].iloc[i])
    diff=df['difficulty_parameters'].iloc[i]
    row=pd.DataFrame(diffparam.iloc[i].dropna())
    row=row.transpose()
    for pos,key in enumerate(diff.keys()):
        data.insert(pos,key,row[key].iloc[0])
    list_df.append(data)
len(list_df)

#list_df variable containts every row of the original dataset expanded as a single dataset.
#To access each dataset, just iterate through the list elements.
```

The Above program was coded to extract 149 different dataframes based on every row as discussed .
But upon my analysis of the dataset, every algorithm has a unique set of difficulty parameters. So i think it would be better if we come up with 6 datasets(One for each algorithm)
But in that case the number of rows in each dataset will be higher.
I have Coded the same below, Just Check once if this will do

```python
def extract(algo):
    df=original_dataset[original_dataset['daa']==algo]
    diffparam=df['difficulty_parameters'].apply(pd.Series)
    df1=df['results'].iloc[0]
    dataframe=pd.DataFrame(df1)
    diff=df['difficulty_parameters'].iloc[0]
    row=pd.DataFrame(diffparam.iloc[0].dropna())
    row=row.transpose()
    for pos,key in enumerate(diff.keys()):
        dataframe.insert(pos,key,row[key].iloc[0])
    for i in range(1,len(df)):
        df1=df['results'].iloc[i]
        data=pd.DataFrame(df1)
        diff=df['difficulty_parameters'].iloc[i]
        row=pd.DataFrame(diffparam.iloc[i].dropna())
        row=row.transpose()
        for pos,key in enumerate(diff.keys()):
            data.insert(pos,key,row[key].iloc[0])
        dataframe=pd.concat([dataframe,data],ignore_index=True)
    dataframe.insert(0,'daa',df['daa'].iloc[0])
    return dataframe
```

```python
algorithms=list(original_dataset['daa'].unique())
dataset=dict()
for algo in algorithms:
    dataset[algo]=extract(algo)
#The Dataset Variable is a dictonary containing dataset extracted for each algorithm with the key as the algorithm's name i.e daa
#Given below is a demo of what each dataset will contain.  
```
